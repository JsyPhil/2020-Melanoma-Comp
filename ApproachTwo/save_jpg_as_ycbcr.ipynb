```python
import pandas as pd
import cv2
import numpy as np
```

```python
'''
Update folders as needed to be able to read and write images to a normalised ycbcr form
---------------------------------------------------------------------------------------
csv_train_file and csv_test_file are the csv files where we read the image from
image_train_folder and image_test_folder are where the jpegs are currently stored
image_train_folder_ycbcr and image_train_folder_ycbcr need to be created to write rescaled YCrCb images
image_train_folder_ycbcr_norm and image_train_folder_ycbcr_norm need to be created to write normalised CrCb vectors in each of the images
'''

csv_train_file = pd.read_csv('/home/malmason/datasets/siim-isic-melanoma-classification/train.csv')
csv_test_file = pd.read_csv('/home/malmason/datasets/siim-isic-melanoma-classification/test.csv')
img_train_folder = '/home/malmason/datasets/siim-isic-melanoma-classification/jpeg/train/'
img_test_folder = '/home/malmason/datasets/siim-isic-melanoma-classification/jpeg/test/'
img_train_folder_ycbcr = '/home/malmason/datasets/siim-isic-melanoma-classification/ycbcr/train/'
img_test_folder_ycbcr = '/home/malmason/datasets/siim-isic-melanoma-classification/ycbcr/test/'
img_train_folder_ycbcr_norm = '/home/malmason/datasets/siim-isic-melanoma-classification/ycbcr/norm/train/'
img_test_folder_ycbcr_norm = '/home/malmason/datasets/siim-isic-melanoma-classification/ycbcr/norm/test/'

X_train_img = csv_train_file['image_name']
X_test_img = csv_test_file['image_name']
```

```python
temp_count = 0
for image_get in X_train_img:
    img_train = cv2.imread(img_train_folder + '{}.jpg'.format(image_get))
    img_train = cv2.resize(img_train,(640,480))
    img_train = img_train[0:480  ,  80:560]
    img_train = cv2.cvtColor(img_train, cv2.COLOR_BGR2YCrCb)
    cv2.imwrite(img_train_folder_ycbcr + '{}.jpg'.format(image_get), img_train)
    if temp_count % 500 == 0:
        print(f'Images processed: {temp_count}')
    temp_count +=1
```

```python
temp_count = 0
for image_get in X_test_img:
    img_test = cv2.imread(img_test_folder + '{}.jpg'.format(image_get))
    img_test = cv2.resize(img_test,(640,480))
    img_test = img_test[0:480  ,  80:560]
    img_test = cv2.cvtColor(img_test, cv2.COLOR_BGR2YCrCb)
    cv2.imwrite(img_test_folder_ycbcr + '{}.jpg'.format(image_get), img_test)
    if temp_count % 500 == 0:
        print(f'Images processed: {temp_count}')
    temp_count +=1
```

```python
temp_count = 0
for image_get in X_train_img:
    img_train_norm = cv2.imread(img_train_folder_ycbcr + '{}.jpg'.format(image_get))

    Y = img_train_norm[:,:,0]
    Cr = img_train_norm[:,:,1]
    Cb = img_train_norm[:,:,2]
    
    Cr_min = np.min(Cr)
    Cr_max = np.max(Cr)
    Cr_range = np.max(Cr) - np.min(Cr)
    Cr = np.subtract(Cr, Cr_min)
    Cr = np.divide(Cr, Cr_range)
    Cr = np.multiply(Cr, 255)

    Cb_min = np.min(Cb)
    Cb_max = np.max(Cb)
    Cb_range = np.max(Cb) - np.min(Cb)
    Cb = np.subtract(Cb, Cb_min)
    Cb = np.divide(Cb, Cb_range)
    Cb = np.multiply(Cb, 255)

    img_ycrcb_norm = np.zeros((480,480, 3), dtype=np.uint8)
    img_ycrcb_norm[:,:,0] = Y
    img_ycrcb_norm[:,:,1] = Cr
    img_ycrcb_norm[:,:,2] = Cb

    cv2.imwrite(img_train_folder_ycbcr_norm + '{}.jpg'.format(image_get), img_ycrcb_norm)
    if temp_count % 1000 == 0:
        print(f'Images processed: {temp_count}')
        print (f'Y_min: {np.min(img_ycrcb_norm[:,:,0])} \tY_max: {np.max(img_ycrcb_norm[:,:,0])}',
               f'\nCr_min: {np.min(img_ycrcb_norm[:,:,1])} \tCr_max: {np.max(img_ycrcb_norm[:,:,1])}'
               f'\nCb_min: {np.min(img_ycrcb_norm[:,:,2])} \tCb_max: {np.max(img_ycrcb_norm[:,:,2])}\n')
    temp_count +=1
```

```python
temp_count = 0
for image_get in X_test_img:
    img_test_norm = cv2.imread(img_test_folder_ycbcr + '{}.jpg'.format(image_get))

    Y = img_test_norm[:,:,0]
    Cr = img_test_norm[:,:,1]
    Cb = img_test_norm[:,:,2]
    
    Cr_min = np.min(Cr)
    Cr_max = np.max(Cr)
    Cr_range = np.max(Cr) - np.min(Cr)
    Cr = np.subtract(Cr, Cr_min)
    Cr = np.divide(Cr, Cr_range)
    Cr = np.multiply(Cr, 255)

    Cb_min = np.min(Cb)
    Cb_max = np.max(Cb)
    Cb_range = np.max(Cb) - np.min(Cb)
    Cb = np.subtract(Cb, Cb_min)
    Cb = np.divide(Cb, Cb_range)
    Cb = np.multiply(Cb, 255)

    img_ycrcb_norm = np.zeros((480,480, 3), dtype=np.uint8)
    img_ycrcb_norm[:,:,0] = Y
    img_ycrcb_norm[:,:,1] = Cr
    img_ycrcb_norm[:,:,2] = Cb

    cv2.imwrite(img_test_folder_ycbcr_norm + '{}.jpg'.format(image_get), img_ycrcb_norm)
    if temp_count % 1000 == 0:
        print(f'Images processed: {temp_count}')
    temp_count +=1
```
